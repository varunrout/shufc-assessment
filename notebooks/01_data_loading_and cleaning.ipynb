## Analysis of Missing Values: District and Ward

### Observation
Both `district` and `ward` columns have exactly **28 missing values** (10.1%), along with `distance_to_bramall_lane_km`. This identical missing count suggests these values are missing for the same set of customers.

### Root Cause
The analysis reveals that **district and ward data is only available for customers in England**:

- **UK administrative geography**: Districts and wards are administrative divisions specific to England. Other UK nations (Scotland, Wales, Northern Ireland) use different administrative structures (e.g., council areas in Scotland, principal areas in Wales).
- **International customers**: Customers from outside the UK naturally don't have English district/ward data.
- **Distance calculation dependency**: Since `distance_to_bramall_lane_km` is calculated from geographic coordinates derived from district/ward, it is also missing for these same customers.

### Breakdown
| Country | District/Ward Available |
|---------|------------------------|
| England | Yes |
| Scotland, Wales, Northern Ireland | No (different admin structure) |
| International | No |

### Decision: Exclude Records with Missing Geographical Features
**We will exclude the 28 customers with missing geographical data** (`district`, `ward`, `distance_to_bramall_lane_km`) from the analysis.

**Rationale:**
- Geographical features are essential for location-based analysis and distance segmentation
- The excluded records represent only ~10% of the customer base
- Retaining only England-based customers ensures consistent data for geographic analysis

````python
# Exclude customers with missing geographical features
customers_df_clean = customers_df.dropna(subset=['district', 'ward', 'distance_to_bramall_lane_km'])

print(f"Original customers: {len(customers_df)}")
print(f"Customers after excluding missing geographical data: {len(customers_df_clean)}")
print(f"Records excluded: {len(customers_df) - len(customers_df_clean)}")
````

## Analysis of Missing Age Values

### Key Finding: Missing Age is Primarily an "Account" Issue

Out of 92 customers with missing age data (after excluding those with missing geographical features):

| Customer Type | Missing Age Count | % of All Tickets Involved |
|---------------|-------------------|---------------------------|
| **Accounts** | 71 | 77.93% |
| **Customers** (individuals) | 21 | 1.86% |

### Interpretation

- **Accounts (71 of 92)**: These represent corporate or business entities, not individuals. Age is **not applicable** for organizations  this is expected missing data, not a data quality issue.
- **Individual Customers (21 of 92)**: Only 21 customers have genuinely missing age data, and they account for just **1.86%** of ticket transactions.

### Implication for Data Handling

1. **Do NOT drop all records with missing age**  this would incorrectly remove ~78% of ticket transactions (the bulk of business data).
2. **Treat "Account" and "Customer" types differently**:
   - **Accounts**: Age is N/A by design  retain these records for all non-age-based analysis
   - **Individual Customers with missing age**: Can be excluded for age-based analysis with minimal impact (1.86% of tickets), or handled separately

### Decision
- Retain all records in `customers_df_clean`
- For age-based analysis, filter to `customer_type == 'Customer'` and exclude missing age values where necessary

````python
# Next step is to validate joins between clean customers and ticket sales data, ensuring that we can still link the majority of ticket transactions to customers with complete geographical data. We will check how many tickets can be linked to customers in the cleaned dataset and assess any potential bias introduced by excluding customers with missing geographical features.

customers_df_clean_ids = customers_df_clean['customer_id'].tolist()
tickets_with_clean_customers = ticket_sales_df[ (ticket_sales_df['purchaser_id'].isin(customers_df_clean_ids)) | (ticket_sales_df['owner_id'].isin(customers_df_clean_ids)) ]
print(f"Tickets that can be linked to customers with complete geographical data: {len(tickets_with_clean_customers)}")
print(f"Percentage of all tickets that can be linked to customers with complete geographical data: {len(tickets_with_clean_customers) / len(ticket_sales_df) * 100:.2f}%")

````

````python
# Data Integrity Checks

# 1. Check for duplicate customer IDs (should not exist)
duplicate_customer_ids = customers_df_clean[customers_df_clean['customer_id'].duplicated()]
print(f"Duplicate customer IDs: {len(duplicate_customer_ids)}")
if len(duplicate_customer_ids) > 0:
    display(duplicate_customer_ids)

# 2. Check for orphan records - tickets with purchaser_id or owner_id not in customers
all_customer_ids = set(customers_df_clean['customer_id'].tolist())

orphan_purchasers = ticket_sales_df[~ticket_sales_df['purchaser_id'].isin(all_customer_ids)]
orphan_owners = ticket_sales_df[~ticket_sales_df['owner_id'].isin(all_customer_ids)]

print(f"\nOrphan tickets (purchaser_id not in customers_df_clean): {len(orphan_purchasers)}")
print(f"Orphan tickets (owner_id not in customers_df_clean): {len(orphan_owners)}")

# Unique orphan IDs
orphan_purchaser_ids = set(orphan_purchasers['purchaser_id'].unique())
orphan_owner_ids = set(orphan_owners['owner_id'].unique())
print(f"Unique orphan purchaser IDs: {len(orphan_purchaser_ids)}")
print(f"Unique orphan owner IDs: {len(orphan_owner_ids)}")

# 3. Check for customers with no ticket transactions (neither purchaser nor owner)
purchaser_ids_in_tickets = set(ticket_sales_df['purchaser_id'].unique())
owner_ids_in_tickets = set(ticket_sales_df['owner_id'].unique())
all_ids_in_tickets = purchaser_ids_in_tickets.union(owner_ids_in_tickets)

customers_without_tickets = customers_df_clean[~customers_df_clean['customer_id'].isin(all_ids_in_tickets)]
print(f"\nCustomers with no ticket transactions: {len(customers_without_tickets)}")
if len(customers_without_tickets) > 0:
    display(customers_without_tickets)
````

````python
# Check if orphaned tickets arise from excluded customers (those with missing geographical data)
excluded_customer_ids = set(customers_df['customer_id'].tolist()) - set(customers_df_clean['customer_id'].tolist())
print(f"Customers excluded during cleaning: {len(excluded_customer_ids)}")

# Check how many orphan IDs are from excluded customers
orphan_purchasers_from_excluded = orphan_purchaser_ids.intersection(excluded_customer_ids)
orphan_owners_from_excluded = orphan_owner_ids.intersection(excluded_customer_ids)

print(f"\nOrphan purchaser IDs from excluded customers: {len(orphan_purchasers_from_excluded)} / {len(orphan_purchaser_ids)}")
print(f"Orphan owner IDs from excluded customers: {len(orphan_owners_from_excluded)} / {len(orphan_owner_ids)}")

# Check if there are any truly missing customers (IDs in tickets but not in original customers_df at all)
all_original_customer_ids = set(customers_df['customer_id'].tolist())
truly_missing_purchasers = orphan_purchaser_ids - all_original_customer_ids
truly_missing_owners = orphan_owner_ids - all_original_customer_ids

print(f"\nTruly missing purchaser IDs (not in original customers_df): {len(truly_missing_purchasers)}")
print(f"Truly missing owner IDs (not in original customers_df): {len(truly_missing_owners)}")

if len(orphan_purchasers_from_excluded) == len(orphan_purchaser_ids) and len(orphan_owners_from_excluded) == len(orphan_owner_ids):
    print("\n All orphan tickets are from customers excluded during cleaning (missing geographical data).")
    print("  No truly missing customer records.")
````

````python
# Filter out orphaned tickets - keep only tickets where BOTH purchaser_id AND owner_id are in cleaned customers
ticket_sales_df_clean = ticket_sales_df[
    (ticket_sales_df['purchaser_id'].isin(all_customer_ids)) & 
    (ticket_sales_df['owner_id'].isin(all_customer_ids))
]

print(f"Original tickets: {len(ticket_sales_df)}")
print(f"Tickets after removing orphans: {len(ticket_sales_df_clean)}")
print(f"Tickets removed: {len(ticket_sales_df) - len(ticket_sales_df_clean)}")
print(f"Percentage retained: {len(ticket_sales_df_clean) / len(ticket_sales_df) * 100:.2f}%")
````

## Data Quality Checks on Cleaned Datasets

Running comprehensive DQ checks on `ticket_sales_df_clean` and `customers_df_clean` to validate data quality before analysis.

````python
# ============================================
# DATA QUALITY CHECKS
# ============================================
from datetime import datetime

dq_results = []

# --- COMPLETENESS CHECKS ---
print("=" * 60)
print("1. COMPLETENESS CHECKS")
print("=" * 60)

# Missing values in cleaned tickets
tickets_missing = ticket_sales_df_clean.isnull().sum()
tickets_missing_total = tickets_missing.sum()
print(f"\nTickets - Missing values: {tickets_missing_total}")
if tickets_missing_total > 0:
    print(tickets_missing[tickets_missing > 0])
dq_results.append(("Completeness", "Tickets - No missing values", tickets_missing_total == 0))

# Missing values in cleaned customers (excluding age for Accounts)
customers_missing = customers_df_clean.isnull().sum()
print(f"\nCustomers - Missing values by column:")
print(customers_missing[customers_missing > 0])

# --- UNIQUENESS CHECKS ---
print("\n" + "=" * 60)
print("2. UNIQUENESS CHECKS")
print("=" * 60)

# Duplicate transaction IDs
duplicate_transactions = ticket_sales_df_clean['transaction_id'].duplicated().sum()
print(f"\nTickets - Duplicate transaction_id: {duplicate_transactions}")
dq_results.append(("Uniqueness", "Tickets - No duplicate transaction_id", duplicate_transactions == 0))

# Duplicate customer IDs
duplicate_customers = customers_df_clean['customer_id'].duplicated().sum()
print(f"Customers - Duplicate customer_id: {duplicate_customers}")
dq_results.append(("Uniqueness", "Customers - No duplicate customer_id", duplicate_customers == 0))

# --- VALIDITY CHECKS ---
print("\n" + "=" * 60)
print("3. VALIDITY CHECKS")
print("=" * 60)

# Date validity: transaction_date should be <= event_date
invalid_dates = ticket_sales_df_clean[ticket_sales_df_clean['transaction_date'] > ticket_sales_df_clean['event_date']]
print(f"\nTickets with transaction_date > event_date: {len(invalid_dates)}")
dq_results.append(("Validity", "Tickets - transaction_date <= event_date", len(invalid_dates) == 0))
if len(invalid_dates) > 0:
    display(invalid_dates.head())

# Age validity: should be between 0 and 120 (for Customers only)
customers_with_age = customers_df_clean[customers_df_clean['age'].notna()]
invalid_ages = customers_with_age[(customers_with_age['age'] < 0) | (customers_with_age['age'] > 120)]
print(f"\nCustomers with invalid age (outside 0-120): {len(invalid_ages)}")
dq_results.append(("Validity", "Customers - Age in valid range (0-120)", len(invalid_ages) == 0))

# Distance validity: should be positive
invalid_distances = customers_df_clean[customers_df_clean['distance_to_bramall_lane_km'] < 0]
print(f"Customers with negative distance: {len(invalid_distances)}")
dq_results.append(("Validity", "Customers - Distance is positive", len(invalid_distances) == 0))

# --- ACCURACY CHECKS ---
print("\n" + "=" * 60)
print("4. ACCURACY CHECKS")
print("=" * 60)

# No future dates
today = pd.Timestamp('2026-02-13')
future_transactions = ticket_sales_df_clean[ticket_sales_df_clean['transaction_date'] > today]
future_events = ticket_sales_df_clean[ticket_sales_df_clean['event_date'] > today]
print(f"\nTickets with future transaction_date: {len(future_transactions)}")
print(f"Tickets with future event_date: {len(future_events)}")
dq_results.append(("Accuracy", "Tickets - No future transaction dates", len(future_transactions) == 0))

# Date range
print(f"\nTransaction date range: {ticket_sales_df_clean['transaction_date'].min()} to {ticket_sales_df_clean['transaction_date'].max()}")
print(f"Event date range: {ticket_sales_df_clean['event_date'].min()} to {ticket_sales_df_clean['event_date'].max()}")

# --- CONSISTENCY CHECKS ---
print("\n" + "=" * 60)
print("5. CONSISTENCY CHECKS")
print("=" * 60)

# All purchaser_id and owner_id exist in customers
clean_customer_ids = set(customers_df_clean['customer_id'].tolist())
orphan_check_purchaser = ticket_sales_df_clean[~ticket_sales_df_clean['purchaser_id'].isin(clean_customer_ids)]
orphan_check_owner = ticket_sales_df_clean[~ticket_sales_df_clean['owner_id'].isin(clean_customer_ids)]
print(f"\nTickets with invalid purchaser_id: {len(orphan_check_purchaser)}")
print(f"Tickets with invalid owner_id: {len(orphan_check_owner)}")
dq_results.append(("Consistency", "Tickets - All purchaser_id in customers", len(orphan_check_purchaser) == 0))
dq_results.append(("Consistency", "Tickets - All owner_id in customers", len(orphan_check_owner) == 0))

# --- DQ SUMMARY ---
print("\n" + "=" * 60)
print("DQ CHECK SUMMARY")
print("=" * 60)
for category, check, passed in dq_results:
    status = " PASS" if passed else " FAIL"
    print(f"{status} | {category}: {check}")
````
