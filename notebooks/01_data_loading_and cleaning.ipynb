## Missing district/ward — it's a geography thing

The 28 missing values for `district`, `ward`, and `distance_to_bramall_lane_km` are all the same customers. Turns out districts and wards are England-specific administrative divisions — customers from Scotland, Wales, NI, or internationally just don't have these fields.

So this isn't dirty data, it's a structural limitation. Distance is derived from district/ward, which is why it's also missing for these 28.

Rather than dropping them (which would bias the dataset toward local fans and lose potentially high-value corporate/international customers), I'll keep everyone and add an `is_england_customer` flag. Distance-based analysis can filter on `is_england_customer == True` where needed.

```python
# flag non-England customers instead of dropping them
customers_df_clean = customers_df.copy()
customers_df_clean['is_england_customer'] = customers_df_clean['district'].notna()

print(f"Total customers: {len(customers_df_clean)}")
print(f"England customers (has district/ward/distance): {customers_df_clean['is_england_customer'].sum()}")
print(f"Non-England customers (missing geographical data): {(~customers_df_clean['is_england_customer']).sum()}")
print(f"\nNon-England customers retained for analysis — geographical features used only where applicable.")
```

```python
# Explore missing age values
missing_age_df = customers_df_clean[customers_df_clean['age'].isnull()]
display(missing_age_df)
```

```python
# break down missing age by customer type — suspecting Accounts don't need age
missing_age_accounts = missing_age_df[missing_age_df['customer_type'] == 'Account']
missing_age_customers = missing_age_df[missing_age_df['customer_type'] == 'Customer']  
print(f"Missing age - Accounts: {len(missing_age_accounts)}")
print(f"Missing age - Customers: {len(missing_age_customers)}")

# how many tickets do these touch?
accounts_missing_age_ids = missing_age_accounts['customer_id'].tolist()
tickets_as_purchaser_accounts = ticket_sales_df[ticket_sales_df['purchaser_id'].isin(accounts_missing_age_ids)]
tickets_as_owner_accounts = ticket_sales_df[ticket_sales_df['owner_id'].isin(accounts_missing_age_ids)]
tickets_involved_accounts = ticket_sales_df[ (ticket_sales_df['purchaser_id'].isin(accounts_missing_age_ids)) | (ticket_sales_df['owner_id'].isin(accounts_missing_age_ids)) ]
print(f"Tickets purchased by accounts with missing age: {len(tickets_as_purchaser_accounts)}")
print(f"Tickets owned by accounts with missing age: {len(tickets_as_owner_accounts)}")
print(f"Total tickets involving accounts with missing age (purchaser OR owner): {len(tickets_involved_accounts)}")
print(f"Percentage of all tickets involving accounts with missing age: {len(tickets_involved_accounts) / len(ticket_sales_df) * 100:.2f}%")

customers_missing_age_ids = missing_age_customers['customer_id'].tolist()
tickets_as_purchaser_customers = ticket_sales_df[ticket_sales_df['purchaser_id'].isin(customers_missing_age_ids)]
tickets_as_owner_customers = ticket_sales_df[ticket_sales_df['owner_id'].isin(customers_missing_age_ids)]
tickets_involved_customers = ticket_sales_df[ (ticket_sales_df['purchaser_id'].isin(customers_missing_age_ids)) | (ticket_sales_df['owner_id'].isin(customers_missing_age_ids)) ]
print(f"Tickets purchased by customers with missing age: {len(tickets_as_purchaser_customers)}")
print(f"Tickets owned by customers with missing age: {len(tickets_as_owner_customers)}")
print(f"Total tickets involving customers with missing age (purchaser OR owner): {len(tickets_involved_customers)}")
print(f"Percentage of all tickets involving customers with missing age: {len(tickets_involved_customers) / len(ticket_sales_df) * 100:.2f}%")

```

## Missing age — mostly Accounts, not a real problem

71 of the 92 missing-age records are Accounts (business entities). Age genuinely doesn't apply to them — they have `gender = Unspecified` too. That's expected, not a quality issue.

Only 21 individual Customers have missing age, and they account for just 1.86% of tickets. So the plan is: keep everything, but filter to `customer_type == 'Customer'` with non-null age when doing age-specific analysis. No imputation needed since the impact is tiny.

## Data quality checks

Running validation on the cleaned datasets before moving to analysis.

```python
# DQ checks on cleaned datasets
from datetime import datetime

dq_results = []

# completeness
print("1. Completeness")
print("-" * 40)

# Missing values in cleaned tickets
tickets_missing = ticket_sales_df_clean.isnull().sum()
tickets_missing_total = tickets_missing.sum()
print(f"\nTickets - Missing values: {tickets_missing_total}")
if tickets_missing_total > 0:
    print(tickets_missing[tickets_missing > 0])
dq_results.append(("Completeness", "Tickets - No missing values", tickets_missing_total == 0))

# Missing values in cleaned customers (excluding age for Accounts)
customers_missing = customers_df_clean.isnull().sum()
print(f"\nCustomers - Missing values by column:")
print(customers_missing[customers_missing > 0])

# --- UNIQUENESS CHECKS ---
# uniqueness
print("\n2. Uniqueness")
print("-" * 40)

# Duplicate transaction IDs
duplicate_transactions = ticket_sales_df_clean['transaction_id'].duplicated().sum()
print(f"\nTickets - Duplicate transaction_id: {duplicate_transactions}")
dq_results.append(("Uniqueness", "Tickets - No duplicate transaction_id", duplicate_transactions == 0))

# Duplicate customer IDs
duplicate_customers = customers_df_clean['customer_id'].duplicated().sum()
print(f"Customers - Duplicate customer_id: {duplicate_customers}")
dq_results.append(("Uniqueness", "Customers - No duplicate customer_id", duplicate_customers == 0))

# --- VALIDITY CHECKS ---
# validity
print("\n3. Validity")
print("-" * 40)

# Date validity: transaction_date should be <= event_date
invalid_dates = ticket_sales_df_clean[ticket_sales_df_clean['transaction_date'] > ticket_sales_df_clean['event_date']]
print(f"\nTickets with transaction_date > event_date: {len(invalid_dates)}")
dq_results.append(("Validity", "Tickets - transaction_date <= event_date", len(invalid_dates) == 0))
if len(invalid_dates) > 0:
    display(invalid_dates.head())

# Age validity: should be between 0 and 120 (for Customers only)
customers_with_age = customers_df_clean[customers_df_clean['age'].notna()]
invalid_ages = customers_with_age[(customers_with_age['age'] < 0) | (customers_with_age['age'] > 120)]
print(f"\nCustomers with invalid age (outside 0-120): {len(invalid_ages)}")
dq_results.append(("Validity", "Customers - Age in valid range (0-120)", len(invalid_ages) == 0))

# Distance validity: should be positive
invalid_distances = customers_df_clean[customers_df_clean['distance_to_bramall_lane_km'] < 0]
print(f"Customers with negative distance: {len(invalid_distances)}")
dq_results.append(("Validity", "Customers - Distance is positive", len(invalid_distances) == 0))

# --- ACCURACY CHECKS ---
# accuracy
print("\n4. Accuracy")
print("-" * 40)

# No future dates
today = pd.Timestamp('2026-02-13')
future_transactions = ticket_sales_df_clean[ticket_sales_df_clean['transaction_date'] > today]
future_events = ticket_sales_df_clean[ticket_sales_df_clean['event_date'] > today]
print(f"\nTickets with future transaction_date: {len(future_transactions)}")
print(f"Tickets with future event_date: {len(future_events)}")
dq_results.append(("Accuracy", "Tickets - No future transaction dates", len(future_transactions) == 0))

# Date range
print(f"\nTransaction date range: {ticket_sales_df_clean['transaction_date'].min()} to {ticket_sales_df_clean['transaction_date'].max()}")
print(f"Event date range: {ticket_sales_df_clean['event_date'].min()} to {ticket_sales_df_clean['event_date'].max()}")

# --- CONSISTENCY CHECKS ---
# consistency
print("\n5. Consistency")
print("-" * 40)

# All purchaser_id and owner_id exist in customers
clean_customer_ids = set(customers_df_clean['customer_id'].tolist())
orphan_check_purchaser = ticket_sales_df_clean[~ticket_sales_df_clean['purchaser_id'].isin(clean_customer_ids)]
orphan_check_owner = ticket_sales_df_clean[~ticket_sales_df_clean['owner_id'].isin(clean_customer_ids)]
print(f"\nTickets with invalid purchaser_id: {len(orphan_check_purchaser)}")
print(f"Tickets with invalid owner_id: {len(orphan_check_owner)}")
dq_results.append(("Consistency", "Tickets - All purchaser_id in customers", len(orphan_check_purchaser) == 0))
dq_results.append(("Consistency", "Tickets - All owner_id in customers", len(orphan_check_owner) == 0))

# --- DQ SUMMARY ---
# summary
print("\nResults")
print("-" * 40)
for category, check, passed in dq_results:
    status = "✓ PASS" if passed else "✗ FAIL"
    print(f"{status} | {category}: {check}")
```

## Cleaning summary

What was done:
- Converted `transaction_date` and `event_date` to datetime
- Confirmed no duplicate customer IDs or transaction IDs
- Added `is_england_customer` flag for the 28 non-England customers (rather than dropping them)
- Kept all records with missing age — Accounts don't need age, and only 21 individual Customers have it missing (1.86% of tickets)
- Validated: no orphan tickets, all ages 0–120, all distances positive, transaction dates don't exceed event dates

Both datasets kept at 100% — 277 customers, 43,289 tickets. Nothing was dropped.

### On non-England customers
Excluding them would bias toward local fans. Instead they're flagged and can be filtered for distance-based work. District/ward/distance fields are England-specific, so those analyses just need `is_england_customer == True`.

### DQ monitoring suggestion
For ongoing data loads, worth implementing automated checks — things like uniqueness on IDs, null thresholds on age (for Customers specifically), referential integrity between tickets and customers. Something like Great Expectations would work well here. The key metrics to track: % missing age for Customers (currently 13.4%), % unmatched ticket IDs (currently 0%), % marketing opt-outs (for reach planning).
